```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
rng=np.random
```

```python
#การจำลองข้อมูล
x = rng.rand(50)*10
y = 2*x+rng.randn(50)
```

```python
#linear regression model
model = LinearRegression()
x_new = x.reshape(-1,1)

print(x_new)
```

```python
#train
model.fit(x_new, y)

print(model.score(x_new,y))
```

```python
#test model
xfit=np.linspace(-1,11)
xfit_new = xfit.reshape(-1,1)

yfit = model.predict(xfit_new)
```

```python
#analysis model
plt.scatter(x,y)
plt.plot(xfit, yfit)
plt.show()
```
