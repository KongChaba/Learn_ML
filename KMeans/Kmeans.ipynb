```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
x,y = make_blobs(n_samples=500,centers=4,cluster_std=0.5,random_state=0)
print(x)
```

```python
print(x[:,0])
```

```python
print(y.shape)
```

```python
model = KMeans(n_clusters=4)
model.fit(x)
y_pred = model.predict(x)
centers = model.cluster_centers_
```

```python
print(centers)
```

```python
#print (centers)
plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.scatter(centers[0,0],centers[0,1],c='red',label="Centroid I")
plt.scatter(centers[1,0],centers[1,1],c='green',label="Centroid II")
plt.scatter(centers[2,0],centers[2,1],c='blue',label="Centroid III")
plt.scatter(centers[3,0],centers[3,1],c='purple',label="Centroid IV")
plt.legend(frameon=True)
plt.show()
```

```python
#new point
x_test, y_test = make_blobs(n_samples=30,centers=4,cluster_std=0.5,random_state=0)
```

```python
y_pred_new = model.predict(x_test)
```

```python
#print (centers)
plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.scatter(x_test[:,0],x_test[:,1],c=y_pred_new,s=90)
plt.scatter(centers[0,0],centers[0,1],c='red',label="Centroid I")
plt.scatter(centers[1,0],centers[1,1],c='green',label="Centroid II")
plt.scatter(centers[2,0],centers[2,1],c='blue',label="Centroid III")
plt.scatter(centers[3,0],centers[3,1],c='purple',label="Centroid IV")
plt.legend(frameon=True)
plt.show()
```
